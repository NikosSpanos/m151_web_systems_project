```python
import json
import matplotlib.pyplot as plt

with open("/home/nspanos/m151_web_systems_project/ml_models/trip_duration/20240527/short_trip/metadata/linear_regressor_scores.json") as f:
    data:dict = json.load(f)

chunks = []
rmse_values = []

for chunk, metrics in data.items():
    if metrics["percentage_improvement"] != 0.0:
        chunks.append(chunk)
        rmse_values.append(metrics["rmse"])

plt.figure(figsize=(10, 6))
plt.plot(chunks, rmse_values, marker='o', linestyle='-', color='b')
plt.xlabel('Chunk')
plt.ylabel('RMSE')
plt.title('RMSE Scores for Chunks with Non-Zero Percentage Improvement')
plt.grid(True)
plt.show()
```
